```python
###HEADERS

get_ipython().magic(u'pylab inline')
import pandas as pd
from scipy import stats
import numpy as np
import scipy as sp


##read in data as a dataframe
import csv
df = pd.read_csv("/Users/nne1s/Documents/Nneka.csv")
print df


```

```python
## data frame with only numbers NOO GENE NAMES
data = df.loc[0:16281, "ASNSF1": "ASNSF4"]
print data
```

```python
###FIND MEAN, STD, COV, MEDIAN OF COV, EXPRESSION CONTROL, LOG BASE VALUES
mean = np.mean(data, axis = 1) #mean
#mean = pd.DataFrame(mean)
std = np.std(data, axis = 1) #standard deviation
cov1 = stats.variation(data, axis = 1) #coeffient of variation
cov = pd.Series(cov1)
median = np.median(cov1) #median of cov
expcon = median / cov #expression control

logvalues =  np.log2(data)
#print logvalues


##Table of all calculations
al = pd.concat([mean, std, cov, expcon, logvalues], axis = 1, ignore_index=True)
al.columns = ["Mean", "Standard Deviation", "Coefficient of Variation",
              "Expression Control", "Log ASNSF1", "Log ASNSF2", "Log ASNSF3", "Log ASNSF4"]
#print "\n\n HERE \n", al

al = al.set_index(df['GeneName'].values)
print al

filename = 'C:/Users/nne1s/Documents/Ede-DATA.csv'
al.to_csv(filename, sep=',')
```

```python
##FIND CORRELATION COEFFICIENT

results = (logvalues.T).corr(method='pearson') #transpose and find correlation
print "\n \n correlation map \n", results

pearsons_df = pd.DataFrame(results.values, columns = df['GeneName'], index = df['GeneName'].values) ## change to dataframe
print "\n \n here \n", pearsons_df

#export to csv
filename = 'C:/Users/nne1s/Documents/Ede-CorrelationMap.csv'
pearsons_df.to_csv(filename, sep=',')
```
